# R Random Generator

## 1/ 
Le generateur doit suivre une loi uniforme.


```r
createHistogram <- function(n){
    val <- sort(runif(n))

    sturges <- 1 + log2(n)

    mini <- val[1]
    maxi <- val[n]

    ifelse(sturges < 5, sturges <- 5, ifelse(sturges > 20 , sturges <- 20, sturges <- round(sturges)))

    mini <- round(mini)
    maxi <- round(maxi)


    hist(val,prob=T,breaks=seq(mini,maxi,abs(maxi-mini)/sturges))
    lines(density(val))
}

createHistogramEffectif <- function(n){
    val <- sort(runif(n))

    sturges <- 1 + log2(n)


    ifelse(sturges < 5, sturges <- 5, ifelse(sturges > 20 , sturges <- 20, sturges <- round(sturges)))

    mini <- round(val[1]- 0.025 * (val[1]-val[n]))
    maxi <- round(val[n]+ 0.025 * (val[1]-val[n]))

    # print(paste(mini,"---",maxi,"---",abs(maxi-mini)/sturges))

    input = 1:n
    multiple_of_sturges = (input %% round(n/sturges)) == 0

    br <- c(mini)

    for(i in input[multiple_of_sturges]){
        br<- c(br,val[i])
    }

    br[length(br)+1] <- maxi

    hist(val,prob=T,breaks=br)
    lines(density(val))
}
```

## 2/

```r
createHistogram(30)
createHistogramEffectif(30)
```

on remarque que dans ce cas la différence du type de classe n'a pas un gros impact sur le résultat

## 3/

```r
createHistogram(30)
createHistogramEffectif(30)
createHistogram(30)
createHistogramEffectif(30)
```

les histogrammes ne sont pas très uniformes, il faut essayé avec plus de valeur

## 4/
avec 1000 tirages :

```r
createHistogram(1000)
createHistogramEffectif(1000)
```

le résultat correspond plus aux attentes. On peux supposer que le générateur aléatoire respecte les specs.
